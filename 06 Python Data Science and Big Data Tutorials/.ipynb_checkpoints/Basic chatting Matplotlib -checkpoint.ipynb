```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,5,7])
plt.show()
```

```python
plt.plot([1,2,3],[2,5,7])
plt.title("Rishav")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```

```python
x = [8,6,2]
y = [23,45,88]
x2 = [56,47,33]
y2 = [5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import style 
style.use('ggplot')
x = [8,6,2]
y = [3,5,8]
x2 = [6,7,3]
y2 = [5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
style.use('ggplot')
x = [8,6,2]
y = [3,5,8]
x2 = [6,7,3]
y2 = [5,8,9]
plt.plot(x,y,label = "first")
plt.plot(x2,y2 , label="second")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x = [8,6,2]
y = [3,5,8]
x2 = [6,7,3]
y2 = [5,8,9]
plt.plot(x,y,label = "first",linewidth=4)
plt.plot(x2,y2 , label="second")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x = [8,6,2]
y = [3,5,8]
x2 = [6,7,3]
y2 = [5,8,9]
plt.scatter(x,y,label = "first",linewidth=4)
plt.hist(x2,y2 , label="second")
plt.legend()
plt.show()
```
