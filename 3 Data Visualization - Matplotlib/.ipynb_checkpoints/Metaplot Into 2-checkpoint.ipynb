```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11) 
y = x**2
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
#axes.plot(x,y)
plt.tight_layout()
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(y,x)
axes[1].set_title('Second')

plt.tight_layout()
```

```python
#FIGURE SIZE AND DPI
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(3,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```
