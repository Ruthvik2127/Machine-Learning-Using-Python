```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First')
axes[1].plot(y,x)
axes[1].set_title('Second')

plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Square")
ax.plot(x,x**3,label="X Cube") 

ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Square")
ax.plot(x,x**3,label="X Cube") 

ax.legend(loc=10)   # loc =1 to 10
```
