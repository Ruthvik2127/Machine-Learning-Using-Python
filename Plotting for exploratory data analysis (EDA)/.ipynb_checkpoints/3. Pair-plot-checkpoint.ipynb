```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
iris = pd.read_csv("iris.csv")
```

## Pair Plot 

#### Dis-advanages:
* Can be used when number of features are high
* Cannot visulize higher dimensional patterns in 3D and 4D
* Only possible to view 2D patterns.


```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(iris,hue="species",size=3);
plt.show()
```

### As Seen Above,The pair plots can be divided into three parts:

* The diagonal plot which showcases the histogram. The histogram allow us to see the PDF/Probability distribution of a single variable

* Upper triangle and lower triangle. which shows us the scatter plot.

* The scatter plots show us the relationship between the features. The upper and lower triangles are the mirror image of each other
