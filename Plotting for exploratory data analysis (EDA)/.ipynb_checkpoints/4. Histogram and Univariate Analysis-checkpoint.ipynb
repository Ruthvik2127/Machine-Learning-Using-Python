```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("iris.csv")
```

```python
sns.FacetGrid(iris,hue="species",size=5) \
    .map(sns.distplot,"petal_length") \
    .add_legend();
    
plt.show();
```

## Observations:
1. As we can see the petal length less than 2 all the points belongs to setosa.
2. And pdf of versicolor and virginica are overlapping in some regions and if we say Patel length > 2 and Patel length < 4.7 then points will belong to versicolor and else virginica. and if we say this then we are going to do some mistakes just because we have chosen 4.7 as the threshold. and we have chosen 4.7 as threshold because as we can see at 4.5 lots of points belongs to versicolor so if we get a flower of Patel length 4.5 then there will be a higher chance that it will belong to versicolor and same as we can see Patel length more than 4.7 there are lots of points which belong to virginica so then there is higher chance that if Patel length is greater than 4.7 then it will be virginica.

## Observations:
1. As we can see we can separate setosa using Patel length below the value of Patel length is equal to 2. And that means this well separable from other class.
2. But we can see the plot between versicolor and virginica there some overlap between them.

```python
sns.FacetGrid(iris,hue="species",size=5) \
    .map(sns.distplot,"petal_width") \
    .add_legend();
    
plt.show();
```

## Observations:
1. Using Patel width setosa are well separable with some very small intersection.
2. And same as Patel length there is some little overlap between versicolor and virginica.
3. So if we compare Patel length and Patel width then using Patel length the setosa are better separable than using Patel width.
#### Note : As we know rule of thumb are the farther these distrubutin are the more better i.e more the specise are saperable
