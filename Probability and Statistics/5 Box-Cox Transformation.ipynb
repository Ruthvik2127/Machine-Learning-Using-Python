```python
from scipy import stats
import pandas as pd
import numpy as np
import pylab 
import matplotlib.pyplot as plt

%matplotlib inline

# Creat dummy arrays, skewed to the left
x = stats.loggamma.rvs(5, size=500) + 5

# How is the distribution for x?
pd.Series(x).hist()
plt.show()
stats.probplot(x, dist="norm", plot=pylab)
pylab.show()

```
